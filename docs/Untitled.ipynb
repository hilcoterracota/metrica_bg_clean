```python
import pandas as pd
import numpy as np
import json
import pymongo
import datetime

myclient = pymongo.MongoClient("192.168.2.1:27017",username="hterra",password="terrac0t@", unicode_decode_error_handler='ignore')
activitywatch_db = myclient["activitywatch"]
aw_watcher_window_db = myclient["activitywatch"]["aw-watcher-window"]
```

```python
data = pd.DataFrame(aw_watcher_window_db.find())
```

```python
nstr = "NAVEGADORES"
data['app'] = data['app'].str.upper()
data['app'] = data['app'].str.replace(".EXE","")
data['app'] = data['app'].str.replace("EXPLORER",nstr)
data['app'] = data['app'].str.replace("IEXPLORER",nstr)
data['app'] = data['app'].str.replace("FIREFOX",nstr)
data['app'] = data['app'].str.replace("MICROSOFTEDGE",nstr)
data['app'] = data['app'].str.replace("CHROME",nstr)
filter = data["app"] == nstr
data = data.where(filter).dropna() 
filter = data["duration"] != 0
data = data.where(filter).dropna() 
sub_data = data
sub_data['timestamp'] =  pd.to_datetime(sub_data['timestamp'])
sub_data['month'] = sub_data['timestamp'].apply(lambda x: x.month)
filter = sub_data["month"] == 2
sub_data = sub_data.where(filter).dropna() 
filter = sub_data["title"] != ""
sub_data = sub_data.where(filter).dropna() 

result = sub_data.groupby(['title'], as_index=False)['duration'].sum().sort_values(by=['duration'], ascending=False)[1:50]
result['time'] = result['duration'].apply(lambda x:
    '{:.0f} hr {:.0f} min {:.0f} secs'.format(
        divmod(x,60*60)[0],
        *divmod(divmod(x,60*60)[1],60))
                                 )

result

```
